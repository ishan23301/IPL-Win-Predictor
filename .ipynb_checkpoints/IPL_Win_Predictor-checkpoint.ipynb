```python
import pandas as pd
import numpy as np
```

```python
match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
match.head()
```

```python
delivery.head()
```

Columns that are required for our analysis are 
1) Match_ID
2) Batting team 
3) Bowling team 
4) City 
5) Runs left 
6) Balls left
7) Wickets left
8) Target 
9) Current run rate
10) Required run rate
11) Result


```python
runs = delivery.groupby(['match_id','inning']).sum('total_runs')['total_runs'].reset_index()
```

```python
runs 
```

```python
runs = runs[runs.inning == 1]
```

```python
match_df = pd.merge(match, runs, left_on = 'id', right_on = 'match_id')
```

```python
match_df.shape
```

```python
match_df.head()
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
```

```python
teams = [
    'Kolkata Knight Riders',
    'Sunrisers Hyderabad',
    'Kings XI Punjab',
    'Royal Challengers Bangalore',
    'Mumbai Indians',
    'Delhi Capitals',
    'Chennai Super Kings',
    'Rajasthan Royals'
]
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df = match_df[match_df['dl_applied']==0]
```

```python
match_df = match_df[['match_id', 'city', 'total_runs', 'winner']]
```

```python
match_df['total_runs'] = match_df['total_runs']+1
```

```python
match_df.shape
```

```python
match_df.rename(columns = {'total_runs':'target'}, inplace = True)
```

```python
match_df.head()
```

```python
delivery_df = pd.merge(delivery, match_df, on = 'match_id')
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df['batting_team'] = delivery_df['batting_team'].str.replace('Delhi Daredevils', 'Delhi Capitals')
delivery_df['bowling_team'] = delivery_df['bowling_team'].str.replace('Delhi Daredevils', 'Delhi Capitals')
delivery_df['batting_team'] = delivery_df['batting_team'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
delivery_df['bowling_team'] = delivery_df['bowling_team'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
```

```python
delivery_df.shape
```

```python
delivery_df.head()
```

```python
 delivery_df['current score'] = delivery_df.groupby('match_id').cumsum()['total_runs']
```

```python
delivery_df['runs left'] = delivery_df['target'] - delivery_df['current score']
```

```python
delivery_df['balls left'] = 126 - delivery_df['over']*6 - delivery_df['ball']
```

```python
delivery_df.head()
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna('0')
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: x if x == '0' else '1')
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
```

```python
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
```

```python
delivery_df['wickets remaining'] = 10 - wickets
```

```python
delivery_df.head()
```

```python
delivery_df['current rr'] = (delivery_df['current score']/(120-delivery_df['balls left']))*6
```

```python
delivery_df['required rr'] = (delivery_df['runs left']/(delivery_df['balls left']))*6
```

```python
delivery_df.tail()
```

```python
delivery_df = delivery_df[~delivery_df['runs left']<0]
```

```python
def result(df):
    return 1 if df['batting_team'] == df['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result, axis=1)
```

```python
delivery_df.head()
```

```python
final_df = delivery_df[['batting_team', 'bowling_team', 'city','target', 'runs left', 'balls left', 'wickets remaining', 'current rr', 'required rr', 'result']]
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df = final_df[~(final_df['required rr']<0)]
```

```python
final_df = final_df[~(final_df['balls left']==0)]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = final_df.drop(['result'], axis = 1)
y = final_df['result']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.2, random_state = 14)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(final_df[['batting_team','bowling_team','city']])
```

```python
ohe.categories_
```

```python
column_transform = make_column_transformer((OneHotEncoder(categories = ohe.categories_),['batting_team','bowling_team','city']), remainder = 'passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
pipe = make_pipeline(column_transform, lr)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
def match_progress(x_df, match_id, pipe):
    match = x_df[x_df['match_id'] == match_id]
    match = match[(match['ball'] == 6)]
    winner = match['winner'].values[0]
    temp_df = match[['batting_team', 'bowling_team', 'city', 'target', 'runs left', 'balls left', 'wickets remaining', 'current rr', 'required rr']].dropna()
    temp_df = temp_df[temp_df['balls left'] != 0]
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)
    target = temp_df['target'].values[0]
    runs = list(temp_df['runs left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_in_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wickets remaining'])
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    temp_df['wickets_in_over'] = np.array(new_wickets)[:-1] -  np.array(wickets)

    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_in_over','wickets_in_over','lose','win']]
    return temp_df
```

```python
temp_df = (match_progress(delivery_df, 87, pipe))
```

```python
def dataframe(batting_team, bowling_team, city, target, runs_left, balls_left, wickets_left, crr, rrr):
    data = [[batting_team, bowling_team, city, target, runs_left, balls_left, wickets_left, crr, rrr]]
    x = pd.DataFrame(data, columns = ['batting_team', 'bowling_team', 'city', 'target', 'runs left', 'balls left', 'wickets remaining', 'current rr', 'required rr' ])
    return x
```

```python
abbreviation = {
    'csk':'Chennai Super Kings',
    'dc':'Delhi Capitals',
    'mi':'Mumbai Indians',
    'rcb':'Royal Challengers Bangalore',
    'kxip':'Kings XI Punjab',
    'rr':'Rajasthan Royals',
    'srh':'Sunrisers Hyderabad',
    'kkr':'Kolkata Knight Riders'
}
```

```python
batting_team = (input("Enter batting team: ")).lower()
bowling_team = (input("Enter bowling team: ")).lower()
city = input("Enter venue: ")
target = int(input("Enter target: "))
runs_left = int(input("Enter runs left to chase: "))
balls_left = int(input("Enter balls remaining: "))
wickets_left = int(input("Enter wickets remaining: "))
crr = float(((target-runs_left)*6)/(120-balls_left))
rrr = float((runs_left*6)/(balls_left))
x = dataframe(abbreviation[batting_team], abbreviation[bowling_team], city, target, runs_left, balls_left, wickets_left, crr, rrr)
losing_prob = pipe.predict_proba(x)[0][0]*100
winning_prob = pipe.predict_proba(x)[0][1]*100
print("Probability of loss of " + batting_team + ": ", losing_prob)
print("Probability of win of " + batting_team + ": ", winning_prob)



    
```
